## Regression Model Evaluation

```python
column_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
data = pd.read_csv('housing.csv', header=None, delimiter=r"\s+", names=column_names)
```

```python
"""
CRIM - per capita crime rate by town
ZN - proportion of residential land zoned for lots over 25,000 sq.ft.
INDUS - proportion of non-retail business acres per town.
CHAS - Charles River dummy variable (1 if tract bounds river; 0 otherwise)
NOX - nitric oxides concentration (parts per 10 million)
RM - average number of rooms per dwelling
AGE - proportion of owner-occupied units built prior to 1940
DIS - weighted distances to five Boston employment centres
RAD - index of accessibility to radial highways
TAX - full-value property-tax rate per $10,000
PTRATIO - pupil-teacher ratio by town
B - 1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town
LSTAT - % lower status of the population
MEDV - Median value of owner-occupied homes in $1000's"""
```

```python
data
```

## 1. Split this data set into training (80%) and testing (20%) sets.

The `MEDV` field represents the median value of owner-occupied homes (in $1000's) and is the target variable that we will want to predict.

```python
# Your code here :
from sklearn.model_selection import train_test_split
x=data.drop('MEDV',axis=1)
y=data['MEDV']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

## 2. Train a `LinearRegression` model on this data set and generate predictions on both the training and the testing set.

```python
# Your code here :
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
#make predictions
pred=model.predict(x_train)
pred2=model.predict(x_test)
print(pred)
print(pred2)
```

## 3. Calculate and print R-squared for both the training and the testing set.

```python
# Your code here :
from sklearn.metrics import r2_score
R_sq_train=r2_score(y_train,pred)
R_sq_test=r2_score(y_test,pred2)
print(f"R-squared score for training: {R_sq_train}")
print(f"R-squared score for testing: {R_sq_test}")
```

## 4. Calculate and print mean squared error for both the training and the testing set.

```python
# Your code here :
from sklearn.metrics import mean_squared_error
mse_train=mean_squared_error(y_train,pred)
mse_test=mean_squared_error(y_test,pred2)
print(f"mean squared error for training: {mse_train}")
print(f"mean squared error for testing: {mse_train}")



```

## 5. Calculate and print mean absolute error for both the training and the testing set.

```python
# Your code here :
from sklearn.metrics import mean_absolute_error
mae=mean_absolute_error(y_train,pred)
mae2=mean_absolute_error(y_test,pred2)
print(f"mean absolute error for training: {mae}")
print(f"mean absolute error for testing: {mae2}")

```

## 6. Split this data set into training (80%) and testing (20%) sets.

The `class` field represents the type of flower and is the target variable that we will want to predict.

```python
# Your code here :

x_train,x_test,y_train,y_test=train_test_split(df,target,test_size=0.2,random_state=42)

```

## 7. Train a `LogisticRegression` model on this data set and generate predictions on both the training and the testing set.

```python
# Your code here :
from sklearn.linear_model import LogisticRegression
model2=LogisticRegression()
model2.fit(x_train,y_train)
```

```python
#make predictions
pre=model2.predict(x_train)
pre2=model2.predict(x_test)
print(pre)
print(pre2)
```

## 8. Calculate and print the accuracy score for both the training and the testing set.

```python
# Your code here :
from sklearn.metrics import accuracy_score
acct=accuracy_score(y_train,pre)
accs=accuracy_score(y_test,pre2)
print(f"accuracy_scorer for training: {acct}")
print(f"accuracy_score error for testing: {accs}")
```

## 9. Calculate and print the balanced accuracy score for both the training and the testing set.

```python
# Your code here :
from sklearn.metrics import balanced_accuracy_score
bt=balanced_accuracy_score(y_train,pre)
bs=balanced_accuracy_score(y_test,pre2)
print(f"balanced accuracy_scorer for training: {bt}")
print(f"balanced accuracy_score  for testing: {bs}")
```

## 10. Calculate and print the precision score for both the training and the testing set.

```python
# Your code here :
from sklearn.metrics import precision_score
pct=precision_score(y_train,pre,average=None)
pcs=precision_score(y_test,pre2,average=None)
print(f"precision_scorefor training: {pct}")
print(f"precision_score  for testing: {pcs}")
```

## 11. Calculate and print the recall score for both the training and the testing set.

```python
# Your code here :
from sklearn.metrics import recall_score
r1=recall_score(y_train,pre,average=None)
r2=recall_score(y_test,pre2,average=None)
print(f"recall_score for training: {r1}")
print(f"recall_score  for testing: {r2}")

```

## 12. Calculate and print the F1 score for both the training and the testing set.

```python
# Your code here :
from sklearn.metrics import f1_score
f1=f1_score(y_train,pre,average=None)
f2=f1_score(y_test,pre2,average=None)
print(f"f1_score for training:\n {f1}")
print(f"f1_score  for testing:\n {f2}")

```

## 13. Generate confusion matrices for both the training and the testing set.

```python
# Your code here :
from sklearn.metrics import confusion_matrix
c1=confusion_matrix(y_train,pre,)
c2=confusion_matrix(y_test,pre2,)
print(f"confusion_matrix for training:\n {c1}")
print(f"confusion_matrix  for testing:\n {c2}")

```
